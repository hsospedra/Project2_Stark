H0 (Null):
Number of accounts has no influence on client balance.
(Clients with ≥3 accounts do NOT have higher balances.)

H1 (Alternative):
Clients with 3 or more accounts tend to have higher balances.

```python
df_clean = df_merged.copy()
```

```python
df_clean = df_clean[['client_id', 'num_accts', 'bal']]
df_clean
```

```python
df_clean.dropna() #delete missing value
df_clean[df_clean['bal'] >= 0]
```

```python
# Aggregate client-level
client_df = (
    df_clean
    .groupby('client_id', as_index=False)
    .agg({
        'num_accts': 'max',   # account of client
        'bal': 'mean'         # balance mean
    })
)
client_df
# We need one row per client
```

```python
#Creat 2 group client less 3 account, more 3 account
client_df['acct_group'] = np.where(
    client_df['num_accts'] >= 3,
    '3_or_more',
    'less_than_3'
)
client_df['acct_group']
```

```python
# compare balance between 2 group
bal_3plus = client_df[client_df['acct_group'] == '3_or_more']['bal']
bal_3plus.describe()
```

```python
bal_less3 = client_df[client_df['acct_group'] == 'less_than_3']['bal']
bal_less3.describe()
```

```python
#limit very_high threshold
threshold = client_df['bal'].quantile(0.9)
client_df['very_high_bal'] = client_df['bal'] >= threshold
```

```python
df_ratio = pd.crosstab(
    client_df['acct_group'],
    client_df['very_high_bal'],
    normalize='index'
)
df_ratio
```

```python
import matplotlib.pyplot as plt

ax = df_ratio.plot(kind='bar', stacked=True, figsize=(8,5), color=['#A6CEE3', '#1F78B4'])

plt.ylabel('Proportion')
plt.xlabel('Account group')
plt.title('Proportion of Very High Balances by Account Group')
plt.legend(title='Very High Balance')
plt.ylim(0,1)
plt.show()
```

```python
#Chi_square test
contingency = pd.crosstab(
    client_df['acct_group'],
    client_df['very_high_bal']
)

chi2, p, dof, expected = stats.chi2_contingency(contingency)

chi2, p
```

Clients with three or more accounts show a 22.9% proportion of very high balances.
In contrast, clients with fewer than three accounts have only 6.5% in the very high balance category.
=> This indicates that clients holding multiple accounts are significantly more likely to have very high balances.

The proportion of very high balances in the 3_or_more group is approximately 3.5 times higher than in the less_than_3 group.
Most clients in the less_than_3 group remain in the non–very-high balance category (93.5%).

=> There is strong evidence of concentration of very high balances among clients with multiple accounts.

#4. Test clients show higher post-experiment balances or faster balance growth than Control clients (or the inverse)

Hypothesis A
H0: Average balance of Test = Control
H1: Average balance of Test ≠ Control

```python
df_test[['Variation', 'bal']].head()
```

```python
df_control[['Variation', 'bal']].head()
```

```python
df_average_bal = df_merged[df_merged['Variation'].isin(['Test', 'Control'])].copy()
df_average_bal
```

```python
#clean 'bal', notna: values nan, delete bal < 0
df_average_bal = df_average_bal[df_average_bal['bal'].notna()]
df_average_bal = df_average_bal[df_average_bal['bal'] >= 0]
df_average_bal
```

```python
# Aggregate client-level
client_bal = (
    df_average_bal
    .groupby(['client_id', 'Variation'], as_index=False)
    .agg({'bal': 'mean'})
)
client_bal
#1row = 1client
```

```python
# Selects balance values of client in the Test group
bal_test = client_bal[client_bal['Variation'] == 'Test']['bal']
bal_test
```

```python
bal_control = client_bal[client_bal['Variation'] == 'Control']['bal']
bal_control
```

```python
bal_test.describe()
```

```python
bal_control.describe()
```

The mean is much greater than the median in both groups,
indicating a strong right-skewed distribution.
There are very large outliers, with balances reaching millions.
The Control group has a slightly higher mean balance than the Test group; however,
The difference is very small relative to the variability of the data.
=>
A standard (Student’s) t-test should not be used.
A Welch’s t-test is appropriate.

```python
# Hypothesis testing
from scipy import stats

t_stat, p_value = stats.ttest_ind(bal_test, bal_control, equal_var=False)
t_stat, p_value
```

```python
alpha = 0.05

print("Mean Test:", bal_test.mean())
print("Mean Control:", bal_control.mean())

if p_value < alpha:
    print("Reject H0: There is a statistically significant difference in balances.")
else:
    print("Fail to reject H0: No statistically significant difference in balances.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

box = plt.boxplot(
    [bal_test, bal_control],
    labels=['Test', 'Control'],
    showfliers=False,
    patch_artist=True)

colors = ['lightblue', 'lightgreen']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Balance')
plt.title('Balance Distribution (Outliers Hidden)')
plt.show()
```

- Although the Control group shows a slightly higher average balance than the Test group,
the difference is very small relative to the large variability in balances.
- Statistical testing using Welch’s t-test indicates that the difference is not statistically significant,
suggesting that the experiment did not have a meaningful impact on client balances.

Conclusion Based on the Hypothesis
Hypothesis:
Test clients show higher post-experiment balances than Control clients (or the inverse).

Conclusion:
The statistical analysis shows no statistically significant difference in post-experiment balances between Test and Control clients.
Although the Control group has a slightly higher mean balance, the difference is not meaningful given the large variability in the data.
Therefore, the experiment did not demonstrate a measurable impact on client balances.
