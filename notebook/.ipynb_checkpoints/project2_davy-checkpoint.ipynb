# Clients in the test variation are more likely to reach the final confirmation step than clients in the control variation. (does the tes variation improve the conversion funnel?)

```python
# Contingency table 
table = pd.crosstab(df['variation'], df_confirm) 
chi2, p, dof, expected = stats.chi2_contingency(table) 
print("Chi-square:", chi2) 
print("p-value:", p) 
print("Expected frequencies:\n", expected)
```

# The test variation increases engagement (more steps completed) among younger clients compared to older clients.

```python
# Number of steps by visit
engagement = (
    df
    .groupby(['visit_id', 'variation', 'age_group'])
    ['process_step']
    .nunique()
    .reset_index(name='n_steps')
)

engagement
```

```python
engagement['age_binary'] = engagement['age_group'].map({
    'less than 40': '0-40',
    'more than 40': '40+'
})
engagement['age_binary']
```

```python
import statsmodels.formula.api as smf

model = smf.ols(
    formula="""
    n_steps ~ C(variation)
             * C(age_binary)
             + clnt_tenure_yr
             + num_accts
             + bal
             + logons_6_mnth
    """,
    data=engagement
).fit()

model.summary()
```

```python
# H0 : There is no significant relationship between client tenure and average account balances
# with significance level = 0.05

from scipy.stats import pearsonr, spearmanr

# Step 1: Select relevant columns and drop the missing values
df_tenure_balance = df[['clnt_tenure_yr', 'bal']].dropna()

# Step 2: Pearson correlation test
pearson_corr, pearson_p = pearsonr(df_tenure_balance['clnt_tenure_yr'], df_tenure_balance['bal'])

print("Pearson correlation:", pearson_corr)
print("p-value:", pearson_p)

```

p<0 there is a significant relationship between the both variables. So we reject the H0

#### Clients with longer tenure at Vanguard tend to have higher average account balances compared to newer clients
H0 : There is no significant difference between the average account balances of clients with longer tenure compared to newer clients

H1 : There is significant difference between the average account balances of clients with longer tenure compared to newer clients
we will considere that a client with clnt_tenure_yr <= 5 is a new client
alpha = 0.05

```python
from scipy.stats import ttest_ind

# Select relevant columns and drop the missing values
df_tenure_balance = df4[['clnt_tenure_yr', 'bal']].dropna()

# Split the dataset into newer and longer tenuregroups
long_clients_df = df_tenure_balance[df_tenure_balance['clnt_tenure_yr'] > 5]['bal']
new_clients_df = df_tenure_balance[df_tenure_balance['clnt_tenure_yr'] <= 5]['bal']


# Perform statistics 
mean_long = long_clients_df.mean()
std_long = long_clients_df.std(ddof=1)
mean_new = new_clients_df.mean()
std_new = new_clients_df.std(ddof=1)


#perform independent t-test
t_stat, p_value = ttest_ind(long_clients_df, new_clients_df, equal_var=False)

# Print results
print(f"Clients with longer tenure: mean={mean_long:.2f}, std={std_long:.2f}")
print(f"Newer clients: mean={mean_new:.2f}, std={std_new:.2f}")
print(f"T-statistic={t_stat:.2f}, p-value={p_value:.5f}")
```

Longer-tenure clients have significantly higher average balances with the mean of the longer tenure almost the doble of the new
p-value < 0.05 => we reject the H0. and conclude that There is significant difference between the average account balances of clients with longer tenure compared to newer client.
